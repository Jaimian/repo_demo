```python
import pandas as pd
import numpy as np
```

```python
def open_vcf(file_name):
    file_df = pd.read_csv(file_name, sep="\t", comment = "#", header = None)
    # setting up df. Separate the file by tab. Ignore the comment, they start with '#'. I don't what a header
    file_df.columns=["CHROM","POS","ID","REF","ALT","QUAL","FILTER","INFO", "FORMAT", "GROUP_C"]

    return file_df
```

```python
TNBC = open_vcf("TNBC.vcf")
nTNBC = open_vcf("nTNBC.vcf")
HER2 = open_vcf("HER2.vcf")
```

```python
print(TNBC.head(5))
print(nTNBC.head(5))
HER2.head(5)
```

```python
def open_txt(file_name):
    file_df = pd.read_csv(file_name, sep="\t", comment = "#" )
    # setting up df. Separate the file by tab. Ignore the comment, they start with '#'. I don't what a header
    #file_df.columns=["CHROM" ,"START", "STOP", "REFSEQID", "GENESYMBOL", "STRAND"]

    return file_df 
```

```python
hg19 = open_txt("hg19_genes_chr1p_bed.txt")
hg19.head(5)
```

```python
hg19.START[5]
```

```python
hg19.STOP[5]
```

```python
value_list = []
for row in nTNBC.POS:
    gene_sym = None
    for pos in range(len(hg19.START)):
        if row > hg19.START[pos] and row < hg19.STOP[pos]: 
            gene_sym = hg19.GENESYMBOL[pos]       
    value_list.append(gene_sym)
value_list
```

```python
len(value_list)
```

```python
New_s = pd.Series(value_list, name = "GENESYMBOL" )
```

```python
nTNBC.shape
```

```python
final_nTNBC = pd.concat([nTNBC, New_s], axis=1)
```

```python
final_nTNBC.head(50)
```
